```python
import pandas as pd
df = pd.read_csv('clean_dataset.csv')
df.head()
```

```python
X = df.drop('Approved', axis = 1)
y = df['Approved']
```

```python
# Сводная информация о наборе данных

df.info()
```

```python
categorical_columns = ["Industry", "Ethnicity", "Citizen"]
for col in categorical_columns:
  print(f'{col}: {df[col].nunique()}')
```

```python
from sklearn.preprocessing import LabelEncoder

# Преобразуем каждый категориальный столбец
categorical_columns = ["Industry", "Ethnicity", "Citizen"]
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
```

```python
X.head()
```

```python
# Посмотреть соответствие значений для столбца Industry
industry_mapping = dict(enumerate(label_encoders["Industry"].classes_))
print(industry_mapping)
# Посмотреть соответствие значений для Ethnicity
ethnicity_mapping = dict(enumerate(label_encoders["Ethnicity"].classes_))
print(ethnicity_mapping)

# Посмотреть соответствие значений для Citizen
citizen_mapping = dict(enumerate(label_encoders["Citizen"].classes_))
print(citizen_mapping)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X, y) # обучаем модель
```
